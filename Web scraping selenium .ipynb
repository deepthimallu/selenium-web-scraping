### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name,experience_required. You have to scrape first 10 jobs data

### Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location,company_name, full job-description. You have to scrape first 10 jobs data.



```python
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#searching for naukri job titles
url='https://www.naukri.com/'
#getting url
driver.get(url)
```

```python
#searching for the data scientists job position
search_job=driver.find_element_by_id("qsb-keyword-sugg")
search_job
```

```python
#Sending the keys for the position
search_job.send_keys('Data Scientist')
```

```python
search_loc=driver.find_element_by_id("qsb-location-sugg")
search_loc.send_keys('Bangalore')
```

```python
#searching for the search button
search_btn=driver.find_element_by_class_name('btn')
search_btn.click()
```

```python
#Extracting the job_titles of top 10 companies
job_titles=[]
title_tags=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
for i in title_tags:
    title=i.text
    job_titles.append(title)
                                             
job_titles[:10]
```

```python
#Extracting the job location of top 10 companies
job_location=[]
location_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location_tags:
    location=i.text
    job_location.append(location)
job_location[:10]
```

```python
#Extracting top 10 company names
company_names=[]
company_details=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
for i in company_details:
    companies=i.text
    company_names.append(companies)
company_names[:10]
```

```python
#Fetching URL'S
urls=[]
for i in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]'):
    urls.append(i.get_attribute('href'))
len(urls)
```

```python
#Extracting top 10 Companies job description
job_description=[]
for i in urls:
    driver.get(i)
    time.sleep(1) 
    try:
        description=driver.find_element_by_xpath('//section[@class="job-desc"]')
        job_description.append(description.text)
    except NoSuchElementException:
        job_description.append('-')
len(job_description)
```

```python
#Creating the DataFrame
#importing libraries
import pandas as pd
import numpy as np
jobs=pd.DataFrame({'Job_titles':job_titles,'Job_location':job_location,'Company_names':company_names,'Job_description':job_description},index=range(0,20,1))
jobs[:10]
```
